```python
# Загружаем датафрейм, используем первые 3 столбца в качестве MultiIndex('year' 'month', 'day'), отбрасываем последний столбец.
df = pd.read_csv('stockholm_td_adj.dat', header=None, sep=r'\s+', index_col=[0, 1, 2], 
                 usecols=range(6), 
                 names=['year', 'month', 'day', 't1', 't2', 't3'])

df['t'] = df.mean(axis=1) # Усредняем температуру (изначально дано 3 значения)
df.drop(columns=df.columns[:-1], inplace=True) # Удаляем ненужные значения температуры

display(df) # Отображаем полученный датафрейм
```

```python
# Группируем данные по месяцу, дню. После чего усредняем значение температуры, разворачиваем таблицу:
# из MultiIndex('month', 'day') в Index('month'), columns = MultiIndex('t', 'day'). После чего транспонируем, 
# чтобы упростить отображение графиков
d = df.groupby(['month', 'day']).mean().unstack().transpose() 
d.columns = map(lambda x: calendar.month_name[x], d.columns) # Задаем названия месяцев: 1 -> January, ...
d.index = range(1, 32) # MultiIndex('t', 'day') -> Index('day')
display(d) # Отображаем полученный датафрейм
```

```python
ax = d.plot(figsize=(20, 12), linewidth=3) # Создаем объект matplotlib.axes.Axes с необходимыми графиками
ax.legend(loc='center left', bbox_to_anchor=(1.04, 0.5), title='Month', fontsize=18, title_fontsize=20) # Задаём легенду
ax.set_title("Stockholm's average temperature", fontsize=20) # Задаём заглавие рисунка
ax.set_xlabel('Day', fontsize=18) # Подписываем ось абсцисс
ax.set_ylabel(r'$t, ^oC$', fontsize=18); # Подписываем ось ординат, блокируем вывод
```
